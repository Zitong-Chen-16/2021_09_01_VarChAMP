## Recreate and save plate platemaps or batch metadata 

Here we read all metadata information we have and incorporate any other metadata information for each batch of data and save it in a standard format 

* Save folder:
  - s3://imaging-platform/projects/2017_10_19_Profiling_rare_ORFs/workspace/metadata/reprocessed

```python
import pandas as pd
import numpy as np
save_reprocessed_path='/home/ubuntu/bucket/projects/2017_10_19_Profiling_rare_ORFs/workspace/metadata/reprocessed/'
```

### Experiment by Chloe for poscon negcon selection
- 2022_08_22_Batch_1

```python
rootDir='/home/ubuntu/bucket/projects/2017_10_19_Profiling_rare_ORFs/workspace/'
batch='2022_08_22_Batch_1'
platemap=pd.read_csv(rootDir+'/metadata/'+batch+'/platemap/2022_08_08_PPL8_100-800.txt', sep='\t')
platemap['batch']=batch
# batch_platemap= pd.read_csv(rootDir+'/metadata/'+batch+'/platemap/BRD4780_1.txt', sep=" ", header=None)
annot_df=platemap.rename(columns={'plate_map_name':'Metadata_Plate','well_position':'Metadata_Well'})

list_of_control_names=['negcon']
annot_df['control']=False
annot_df.loc[annot_df['control_type'].isin(list_of_control_names),'control']=True
annot_df['Variant']=annot_df['allele'].str.upper().str.rstrip('_NA').str.rstrip(' WT').str.replace(' ','_')
# annot_df['Variant']=annot_df['Sample name']
annot_df['Gene']=annot_df['Variant'].apply(lambda x: x.split('_')[0])
annot_df['MT']=annot_df['Variant'].apply(lambda x: x.split('_')[-1])
# annot_df['Metadata_Sample_uniqe']=annot_df['Variant']+'-'+annot_df['Treatment']
annot_df.loc[annot_df['Gene']==annot_df['MT'],'MT']=''

annot_df['Metadata_batch_Plate']=annot_df['batch']+'-'+annot_df['Metadata_Plate']
annot_df['Variant']=annot_df['Variant'].str.replace('_',' ')
annot_df['Metadata_Sample_Unique']=annot_df['Variant']+'-'+annot_df['vTitre'].astype(str)
annot_df.to_csv(save_reprocessed_path+batch+'.csv',index=False)
annot_df.head()
```
